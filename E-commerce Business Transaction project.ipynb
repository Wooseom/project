```python
#validating the type of variables
df['Date']=pd.to_datetime(df['Date']) #converting a sequence of characters into a pandas datetime object
df[['ProductNo','CustomerNo']]=df[['ProductNo','CustomerNo']].astype('object')
```

```python
#splitting the date column to Month, Weekday, WeekdayName
#and creating a TotalPrice column
df['Month'] = df['Date'].dt.month
df['Weekday'] = df['Date'].dt.weekday
df['WeekdayName'] = df['Date'].dt.day_name()
df['Day']=df['Date'].dt.day

df['Amount']=df['Quantity']*df['Price']
```

# Sales Analysis

```python
df = df[df['Date'] <= '2019-11-30']
```

```python
#creating new dataframe "monthly" group by date with frequency monthly
#sum of the values for Amount, Returned with absolute values, count of unique 'TransactionNo'
monthly = df.groupby(pd.Grouper(key='Date',freq='M')).agg(sold=('Amount','sum'),returned=('Amount',lambda x:sum(x[x<0])),
                                                          nunique = ('TransactionNo','nunique'))
#rolling average of 'sold' over a 3 month
monthly['sold_avg'] = monthly['sold'].rolling(window=3).mean()
monthly['returned']=monthly['returned'].abs()
monthly.index=monthly.index.date 
monthly   
```

```python
monthly_sum = monthly[['sold','returned']].sum(axis=1)
#Among the monthly_sum dataframe, extracting 'sold','returned' values for each month 
monthly_pct = monthly[['sold','returned']].div(monthly_sum, axis=0)
```

```python
monthly_sum
```

```python
monthly_pct
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15,10))
monthly[['sold', 'returned']].plot.bar(ax=ax[0])
ax[0].set_ylabel('Revenue (GBP)')
ax[0].set_xlabel('Month')
ax[0].set_title("Monthly sales and returns")
ax[0].grid(axis='y')

monthly_pct.plot.bar(stacked=True, ax=ax[1])
ax[1].set_ylabel('Percentage')
ax[1].set_xlabel('Month')
ax[1].set_title("Monthly relative amounts of sold and returned")
ax[1].grid(axis='y')

plt.subplots_adjust(hspace=1)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(15,7))
ax2 = plt.twinx()
ax1.plot(monthly.index, monthly['sold'], label='Revenue')
ax1.plot(monthly.index, monthly['sold_avg'], label='3 month revenue average')
ax2.bar(monthly.index, monthly['nunique'], width=8, label='Volume', alpha=0.5)

ax1.set_ylabel('Revenue (GBP)')
ax2.set_ylabel('Volume')
ax1.set_xlabel('Month')
plt.title("Monthly Sales")
plt.grid(True)
ax1.legend(loc=(0.03,0.9))
ax2.legend(loc=(0.3,0.9))

plt.show()
```

# Monthly Analysis

```python
#dataframe with quantity greater than 0
df = df[df['Quantity']>0]
#new dataframe 'bydate' 
bydate = df.groupby('Date').agg(
    UniqueTransactions=('TransactionNo', 'nunique'),
    UniqueProductSold=('TransactionNo', 'count'),
    ProductSold=('Quantity', 'sum'),
    Revenue=('Amount', 'sum')
    ).reset_index()
bydate['Day']= bydate['Date'].dt.day
bydate['Weekday'] = bydate['Date'].dt.weekday
bydate['Month'] = bydate['Date'].dt.month
bydate['WeekdayName'] = bydate['Date'].dt.day_name()
bydate
```

```python
byday = bydate.groupby('Day')[['UniqueTransactions', 'UniqueProductSold', 'ProductSold', 'Revenue']].mean()
byday.columns = ['DailyAvgUniqueTransactions', 'DailyAvgUniqueProductSold', 'DailyAvgProductSold', 'DailyAvgRev']
byday = byday.sort_index()
byday
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(10, 7))

# Plot 'DailyAvgRev' 
byday['DailyAvgRev'].plot(ax=ax, kind='line')

ax.set_xlabel('Day of the Month')
ax.set_ylabel('Average Daily Revenue')
ax.set_title('Average Daily Revenue by Day of the Month')

plt.show()
```
